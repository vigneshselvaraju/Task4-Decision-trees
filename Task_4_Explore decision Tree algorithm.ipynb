```python
#Build the model
x=data
y=y
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
print("shape 0f feature training data: ",x_train.shape)
print("shape 0f target training data: ",y_train.shape)
print("shape 0f feature test data: ",x_test.shape)
print("shape 0f target test data: ",y_test.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)

print("shape 0f feature training data: ",x_train.shape)
print("shape 0f target training data: ",y_train.shape)
print("shape 0f feature test data: ",x_test.shape)
print("shape 0f target test data: ",y_test.shape)

#Applying decision trees algorithm
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix

clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
#Accuracy for training data
predict=clf.predict(x_train)
print("Accuracy of training data:",accuracy_score(predict,y_train)*100,"%")
print("Confusion matrix of training data :'\n'",confusion_matrix(predict,y_train))
sns.heatmap(confusion_matrix(predict,y_train),annot=True,cmap='BuGn')
```

```python
#Accuracy for testing data
predict=clf.predict(x_test)
print("Accuracy of training data:",accuracy_score(predict,y_test)*100,"%")
print("Confusion matrix of training data :'\n'",confusion_matrix(predict,y_test))
sns.heatmap(confusion_matrix(predict,y_test),annot=True,cmap='BuGn')
```

```python
plt.figure(figsize=(19,19))
tree.plot_tree(clf,filled=True,rounded=True,proportion=True,node_ids=True,feature_names=dat.feature_names)
plt.show()
```
